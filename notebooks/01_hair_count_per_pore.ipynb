2. ëª¨ë¸ ì˜ˆì¸¡ ë° ë°”ìš´ë”© ë°•ìŠ¤ í¬ë¡­
- íƒì§€ëœ ì˜ì—­ì„ ë°”ìš´ë”© ë°•ìŠ¤ë¡œ ê°ì§€í•˜ì—¬ ê° ëª¨ê³µ ì˜ì—­ì„ ê°œë³„ì ìœ¼ë¡œ í¬ë¡­
- í¬ë¡­ëœ ì´ë¯¸ì§€ëŠ” ì¶”ê°€ì ì¸ í›„ì²˜ë¦¬ ë˜ëŠ” ë¶„ì„ì— í™œìš© ê°€ëŠ¥

```python
# YOLO ëª¨ë¸ ë¡œë“œ
model = YOLO("hair/hair_model8/weights/best.pt")

# ë°ì´í„° ê²½ë¡œ ì„¤ì •
root = Path("datasets/hair/images/Validation")  # ì…ë ¥ ì´ë¯¸ì§€ í´ë”
output_dir = Path("datasets/hair/cropped")  # í¬ë¡­ëœ ì´ë¯¸ì§€ ì €ì¥ í´ë”
output_dir.mkdir(parents=True, exist_ok=True)  # í´ë”ê°€ ì—†ìœ¼ë©´ ìƒì„±

# YOLO ê°ì§€ ë° ë°”ìš´ë”© ë°•ìŠ¤ í¬ë¡­
plt.figure(figsize=(20, 20))
i = 0  # ì„œë¸Œí”Œë¡¯ ì¸ë±ìŠ¤

# í´ë” ë‚´ JPG ì´ë¯¸ì§€ì— ëŒ€í•´ ë°˜ë³µ ìˆ˜í–‰
for path in root.glob("*.JPG"):
    #  YOLO ëª¨ë¸ ì˜ˆì¸¡ ìˆ˜í–‰
    results = model(path)

    #  ê°ì§€ëœ ì´ë¯¸ì§€
    im_array = results[0].plot()
    im = Image.fromarray(im_array[..., ::-1])  # RGB ë³€í™˜

    #  ì›ë³¸ ì´ë¯¸ì§€ ë¡œë“œ
    original_image = Image.open(path)

    #  ê°ì§€ëœ ë°”ìš´ë”© ë°•ìŠ¤ ì¢Œí‘œ (x1, y1, x2, y2)
    boxes = results[0].boxes.xyxy.cpu().numpy()

    # ê°ì§€ëœ ê°ì²´ë³„ë¡œ ë°”ìš´ë”© ë°•ìŠ¤ í¬ë¡­ í›„ ì €ì¥
    for j, (x1, y1, x2, y2) in enumerate(boxes):
        cropped = original_image.crop((x1, y1, x2, y2))
        cropped_path = output_dir / f"{path.stem}_object{j}.jpg"
        cropped.save(cropped_path)

    i += 1
    ax = plt.subplot(3, 4, i)

   ax.axis("off")

plt.tight_layout()
plt.show()
```
