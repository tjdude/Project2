2. 모델 예측 및 바운딩 박스 크롭
- 탐지된 영역을 바운딩 박스로 감지하여 각 모공 영역을 개별적으로 크롭
- 크롭된 이미지는 추가적인 후처리 또는 분석에 활용 가능

```python
# YOLO 모델 로드
model = YOLO("hair/hair_model8/weights/best.pt")

# 데이터 경로 설정
root = Path("datasets/hair/images/Validation")  # 입력 이미지 폴더
output_dir = Path("datasets/hair/cropped")  # 크롭된 이미지 저장 폴더
output_dir.mkdir(parents=True, exist_ok=True)  # 폴더가 없으면 생성

# YOLO 감지 및 바운딩 박스 크롭
plt.figure(figsize=(20, 20))
i = 0  # 서브플롯 인덱스

# 폴더 내 JPG 이미지에 대해 반복 수행
for path in root.glob("*.JPG"):
    #  YOLO 모델 예측 수행
    results = model(path)

    #  감지된 이미지
    im_array = results[0].plot()
    im = Image.fromarray(im_array[..., ::-1])  # RGB 변환

    #  원본 이미지 로드
    original_image = Image.open(path)

    #  감지된 바운딩 박스 좌표 (x1, y1, x2, y2)
    boxes = results[0].boxes.xyxy.cpu().numpy()

    # 감지된 객체별로 바운딩 박스 크롭 후 저장
    for j, (x1, y1, x2, y2) in enumerate(boxes):
        cropped = original_image.crop((x1, y1, x2, y2))
        cropped_path = output_dir / f"{path.stem}_object{j}.jpg"
        cropped.save(cropped_path)

    i += 1
    ax = plt.subplot(3, 4, i)

   ax.axis("off")

plt.tight_layout()
plt.show()
```

3. 모델 평가 및 시각화
- 학습된 모델을 사용하여 새로운 이미지에 대한 예측 수행
- 결과를 그래프로 시각화하여 탐지된 모공 및 머리카락 확인

```python
# best 케이스가 저장된 pt파일로 모델 로드
model = YOLO("/content/탈모/탈모_model3/weights/best.pt")
# 평가 데이터가 저장된 경로
root = Path("/content/탈모/탈모/탈모_data/valid/images")
paths = list(root.glob('*'))

plt.figure(figsize = (20,20))
i = 0

for path in paths[:9]:
  results = model(path) # 모델에 평가 이미지를 예측
  im_array = results[0].plot()  # 결과 객체에서 시각화 기능 사용
  im = Image.fromarray(im_array[..., ::-1])  # RGB PIL image 변환
  i+=1
  ax = plt.subplot(3,3,i)
  ax.imshow(im)
plt.show()
```
