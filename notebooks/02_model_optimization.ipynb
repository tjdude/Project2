1. 모델 변환 (YOLO -> Tensorflow Lite)
- YOLO 모델을 TFLite 포멧으로 변환

```python
from ultralytics import YOLO

model = YOLO("/content/drive/MyDrive/Colab Notebooks/탈모/yolo300.pt")

# export() 함수를 통해 다양한 플랫폼으로 변환 가능
model.export(format='tflite')
```
